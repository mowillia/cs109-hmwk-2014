```python
# using code 
## http://stackoverflow.com/questions/19379295/linear-regression-with-pandas-dataframe
import statsmodels.api as sm
def fit_line2(x, y):
    """Return slope, intercept of best fit line."""
    X = sm.add_constant(x)
    model = sm.OLS(y, X, missing='drop') # ignores entires where x or y is NaN
    fit = model.fit()
    return fit.params[1], fit.params[0] # could also return stderr in each via fit.bse
```

```python
#your code here

# Linear regression Plot definition
def Lin_Reg_Plot(num):
    x = teamsNsal.ix[num].salary
    y = teamsNsal.ix[num].W

    m, b = fit_line2(x, y)
    N = 100 # could be just 2 if you are only drawing a straight line...
    points = np.linspace(x.min(), x.max(), N)

    plt.figure(figsize=(5,5))

    plt.title('Lin. Reg: W vs. Salary for year '+str(num))

    plt.xlabel("Salary") # axis label
    plt.ylabel("Wins") # axis label


    team = teamsNsal.ix[num].set_index('teamID')

    u = team.salary['OAK'] # selects out OAK salary value
    v = team.W['OAK'] # selects out OAK W value
    
    plt.plot(u,v,'bs')
    plt.plot(points, m*points +b,'k-')
    ax = plt.plot(x, y,'r.')
    
    al = plt.gca()
    al.grid(True)

    label_point(x, y, teamsNsal.ix[num].teamID, al)
        
    plt.show()
```

Write a function to merge the `countries` and `income` data sets for any given year. 

```python
"""
Function
--------
mergeByYear

Return a merged DataFrame containing the income, 
country name and region for a given year. 

Parameters
----------
year : int
    The year of interest

Returns
-------
a DataFrame
   A pandas DataFrame with three columns titled 
   'Country', 'Region', and 'Income'. 

Example
-------
>>> mergeByYear(2010)
"""
#your code here

from pandas import merge

def mergeByYear(num):
    
    #renames label as country
    country_income = gdp_per_capita.rename(columns={'gdp pc test':'Country'})
    #merges according to common elements
    continent_gdp = merge(country_income, countries, how = 'inner')
    #renames year column
    continent_gdp.rename(columns={num:'Income'},inplace = True)
    #selects out desired columns
    cont_gdp = continent_gdp[['Country','Income','Region']]

    return cont_gdp
```

Assume you have two list of numbers, X and Y, with distribution approximately normal. X and Y have standard deviation equal to 1, but the average of X is different from the average of Y.  If the difference in the average of X and the average of Y is larger than 0, how does the proportion of X > a compare to the proportion of Y > a? 

Write a function that analytically calculates the ratio of these two proportions: Pr(X > a)/Pr(Y > a) as function of the difference in the average of X and the average of Y. 

**Hint**: Use the `scipy.stats` module for useful functions related to a normal random variable such as the probability density function, cumulative distribution function and survival function. 

**Update**: Assume Y is normally distributed with mean equal to 0. 

Show the curve for different values of a (a = 2,3,4 and 5).

```python
"""
Function
--------
ratioNormals

Return ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a) as function of 
    the difference in the average of X 
    and the average of Y. 

Parameters
----------
diff : difference in the average of X 
    and the average of Y. 
a : cutoff value

Returns
-------
Returns ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a)
    
Example
-------
>>> ratioNormals(diff = 1, a = 2)
"""
#your code here

import scipy.stats

# Function definition
def ratioNormals(diff, a):

    PrXa = 1-scipy.stats.norm(diff, 1).cdf(a)
    PrYa = 1-scipy.stats.norm(0,1).cdf(a)
    
    return PrXa/PrYa
```

```python
#your code here

#Plots
plt.figure(figsize=(8,8))

x = np.arange(0, 4, 0.1) #array of 30 points from 0 to 10\
for i in [2,3,4,5]:
    fx = ratioNormals(x,a = i)
    plt.plot(x, fx, label='Ratio for a = '+str(i))

plt.legend(bbox_to_anchor=(.5, 1))
plt.yscale('log')
plt.xlabel("<X> minus <Y>") # axis label
plt.ylabel("Pr(X > a)/Pr(Y > a)") # axis label
plt.title("Pr(X > a)/Pr(Y > a) for various mean diff")

plt.show()
```

```python
## in order to account for missing elements
from pandas import isnull 

## calculates total population in region
def pop2012_region(sr):
    
    Reg_pop_2012 = pop_income_2012[['Country', 'Population in 2012', 'Region']]
    Tot_pop = Reg_pop_2012.groupby(["Region"], as_index = False).sum()
    x = Tot_pop.set_index('Region').ix[sr]['Population in 2012']

    return x

## calculates the mean income per person in a region
def mean_I_person(sr):
    g = pop_income_2012.set_index('Region').ix[sr]
    
    numerator = 0
    for k in range(0,len(g)):
        income = g.ix[k][1]
        population = g.ix[k][2]
        
        if isnull(income):
            income = 0
        numerator = numerator + income*population
        
    netpop = pop2012_region(sr)
    
    return numerator/netpop
```
