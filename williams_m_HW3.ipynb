All of the statistics are positively correlated with wins, although BB, HR, and 2B (in that order) are the most correlated. To find which statistics to use we can create linear regression models for each combination of statistics.

First we distinguish our training and testing data.

```python
### Training and Testing Data definition

# we use the years before 2002 as the training data
stats02down = stats_mn[stats_mn['yearID']<2002]

# extracts testing data; 
# we use the years after 2002 as the testing data
stats02up = stats_mn[stats_mn['yearID']>=2002]

# Definition of training y values
Wins_train = np.asarray(stats02down['W'])

# Definition of testing y values (i.e. prediction)
Wins_test = np.asarray(stats02up['W'])
```

Then we import the linear regression model 

```python
from sklearn import linear_model
```

We use the linear model module to define a function which outputs the variance score corresponding to the statistics in the list. The closer the variance score (or coefficient of determination http://en.wikipedia.org/wiki/Coefficient_of_determination#Definitions) is to 1 the more accurate the associated model is and the closer it is to zero the worse the model is. Negative values are really bad (they mean the prediction varies more widely from the data than data varies within itself). 

```python
# Outputs variance score according to a fit using the variables in 'lis'

# We use code from
# http://scikit-learn.org/0.11/auto_examples/linear_model/plot_ols.html

def score_func(lis):
    # defining testing variables
    main_stats_test = np.asarray(stats02up[lis])

    # defining training variables
    main_stats_train = np.asarray(stats02down[lis])

    # Create linear regression object
    reg_main = linear_model.LinearRegression()

    # Train the model via training sets
    reg_main.fit(main_stats_train, Wins_train)
    
    # Outputs list and variance score
    # Variance score: close to 1 -> good prediction. close to 0 -> poor prediction.
    return lis, reg_main.score(main_stats_test, Wins_test)
```

Next, we use itertools to define a function which gives us various combinations of the statistics. 

```python
import itertools

# k chooses k elements and l chooses the lth combination of k elements
def var_choice(k,l):
    return list(list(itertools.combinations(['%HR', '%1B', '%2B', '%3B', '%BB'],k))[l])
```

We import the math module and define the combinatorial function. This will be needed later on.

```python
import math

# factorials are cool!
def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)
```

Now, we can print the Variance scores for various combinations of statistics

```python
i = 1 # defines row
for k in range(1,6):
    for l in range(0,nCr(5,k)): 
        lis = var_choice(k,l)
        scores = score_func(lis)
        print i,' Combination:', scores[0], 'Variance Score: ', scores[1]
        i = i+1
```

Although it's a close call (row 27 and row 18 have particularly high values), we see that we need to use all the statistics in order to get the maximum variance score. 

To get the coefficients which define the full_statistics model we need to go back to our linear regression module.

```python
## Linear regression model: Uses all statistics for a prediction of wins. 

# imports linear model module
from sklearn import linear_model

# definition of test data
full_stats_test = np.asarray(stats02up[['%HR','%1B','%2B', '%3B','%BB']])

# definition of training data
full_stats_train = np.asarray(stats02down[['%HR','%1B','%2B', '%3B','%BB']])

# Create linear regression object
reg_full = linear_model.LinearRegression()

# Train the model via training sets
reg_full.fit(full_stats_train, Wins_train)

# The coefficients of linear variables
print('Coefficients: \n', reg_full.coef_)

# The mean square error; smaller values means better predictions
print("Residual sum of squares: %.2f"
      % np.mean((reg_full.predict(full_stats_test) - Wins_test) ** 2))

# Variance score: 1 is perfect prediction. 0 is poor prediction.
print('Variance score: %.2f' % reg_full.score(full_stats_test, Wins_test))
```

So our linear regression model will be defined by 'reg_full' and will use all five player stats. We define it in terms of a function below. 

```python
# returns offensive predicted wins for baseball stats frame

def OPW(df):
    return reg_full.predict(df[['%HR','%1B','%2B', '%3B','%BB']])
```

Examine the data further by looking at the projections to the first two principal components of the data. #### Problem 2(c) 

In the lecture we discussed how to use cross validation to estimate the optimal value for $k$ (the number of nearest neighbors to base the classification on). Use ***ten fold cross validation*** to estimate the optimal value for $k$ for the iris data set. 

**Note**: For your convenience sklearn does not only include the [KNN classifier](http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html), but also a [grid search function](http://scikit-learn.org/stable/modules/generated/sklearn.grid_search.GridSearchCV.html#sklearn.grid_search.GridSearchCV). The function is called grid search, because if you have to optimize more than one parameter, it is common practice to define a range of possible values for each parameter. An exhaustive search then runs over the complete grid defined by all the possible parameter combinations. This can get very computation heavy, but luckily our KNN classifier only requires tuning of a single parameter for this problem set. 

```python
# imports KNN classifier module 
from sklearn.neighbors import KNeighborsClassifier

# Defines K-Nearest Neighbor Classifier
# we choose a particular k, but it really doesn't matter for the next part
knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=7, weights = 'uniform')
```

```python
# Imports grid search module
from sklearn import grid_search

# Defines range of k values
k_range = list(xrange(1,101,1))

# Defines K-Nearest Neighbor GridSearch on Training data
# Cross validation is 10 fold (cv=10)
knngs_train = grid_search.GridSearchCV(cv = 10, estimator = knn, param_grid = dict(n_neighbors=k_range),n_jobs=1)

# Fits the Grid Search according to the training data
knngs_train.fit(X_train,Y_train);
```

```python
# selects n_neighbor value with maximum average validation score
print 'Optimal Choice of Parameter:', knngs_train.best_params_, '\n'

# Shows Cross validation Score corresponding to Best Paramater
k_best = knngs_train.best_params_.values()[0]
print 'Cross Validation Scores for Optimal Parameter: ', knngs_train.grid_scores_[k_best].cv_validation_scores, '\n'

# Shows score corresponding to above n_neighbor value
print 'Optimal Score:', knngs_train.best_score_, '\n'
```

#### Answer:

So for our test_size =.33 and a random_state = 42, we find that the optimal k value is 5.

#### Problem 3(c) 

Write a **ten-fold cross validation** to estimate the optimal value for $k$ for the digits data set. *However*, this time we are interested in the influence of the number of dimensions we project the data down as well. 

Extend the cross validation as done for the iris data set, to optimize $k$ for different dimensional projections of the data. Create a boxplot showing test scores for the optimal $k$ for each $d$-dimensional subspace with $d$ ranging from one to ten. The plot should have the scores on the y-axis and the different dimensions $d$ on the x-axis. You can use your favorite plot function for the boxplots. [Seaborn](http://web.stanford.edu/~mwaskom/software/seaborn/index.html) is worth having a look at though. It is a great library for statistical visualization and of course also comes with a [`boxplot`](http://web.stanford.edu/~mwaskom/software/seaborn/generated/seaborn.boxplot.html) function that has simple means for changing the labels on the x-axis.

```python
# Defines function which outputs optimal k value and
# corresponding 10 fold cross validation scores
# for a given dimensional reduction of the data

def Scores_dim(dim):
    # Gets first two components of Truncated SVD 
    svd_dim = TruncatedSVD(n_components = dim, random_state = 42)

    # Fits SVD to A
    svd_dim.fit(A_train, B_train)

    # Projects data down to dimensions
    A_train_dim = svd_dim.transform(A_train)

    # Defines K-Nearest Neighbor Classifier
    # we choose a particular k, but it really doesn't matter for the next part
    knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=7, weights = 'uniform')

    # Defines range of k values
    k_range = list(xrange(1,101,1))

    # Defines K-Nearest Neighbor GridSearch on Training data
    # Cross validation is 10 fold (cv=10)
    knngs_train = grid_search.GridSearchCV(cv = 10, estimator = knn, param_grid = dict(n_neighbors=k_range),n_jobs=1)

    # Fits the Grid Search according to the training data
    knngs_train.fit(A_train_dim,B_train);
    
    # selects best parameter
    k_best = knngs_train.best_params_.values()[0]
    
    # validation scores for best parameter
    val_scores = knngs_train.grid_scores_[k_best].cv_validation_scores
    
    return k_best, val_scores
```

```python
# Optimal k value for 4 dimensions
print 'Optimal k value for 4 dimensions: ', Scores_dim(4)[0], '\n'
print '10 Fold Cross Validation Scores for 4 dim: ', Scores_dim(4)[1], '\n'
```

```python
# cross validation score for each dimension
data_scores = [Scores_dim(i)[1] for i in range(1,11) ]
```
