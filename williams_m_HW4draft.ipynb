#### Problem 2(a)

Assume you have $M$ polls with sample sizes $n_1, \dots, n_M$. If the polls are independent, what is the average of the variances of each poll if the true proportion is $p$?

** Your answer here: **

We assume we have $M$ independent polls each of size $n_1,\ldots, n_M$. We assume each poll conforms to a binomial distribution with probability $p$ and number of trials $n_k$. From each poll, we find a number of votes and the standard error of this proportion of votes from the true probability $p$ is $s_k = \sqrt{p(1-p)/n_k}$. The average of the variances (computed from the standard error) for each poll is therefore
$$ \text{Average of Variances} = \frac{1}{M}\sum_{k=1}^{M} s_{k}^2= \frac{p(1-p)}{M}\sum_{k=1}^{M}\frac{1}{n_k} $$.

```python
## Definition of function used to compute theoretical standard
## deviation values


def std_th_calc(num):
    state_info = info_cut[num]
    if state_info['estimates'][0]['party'] == 'Rep':
        p = state_info['estimates'][0]['value']/100 
    elif state_info['estimates'][1]['party'] == 'Rep':
        p = state_info['estimates'][1]['value']/100 
    else:
        p = 0

    # finding the list of observations 
    state_slug = slug_list_cut[num]
    obs_lis = list(polls[state_slug]['Number of Observations'])

    # computing the theoretical standard deviation
    M = len(obs_lis)
    var_th = (1./M)*p*(1-p)*sum(1./np.array(obs_lis))
    return np.sqrt(var_th)
```

# Problem 3: Prediction and Posterior Probabilities

Perform a Bayesian analysis to predict the probability of Republicans winning in each state then provide a posterior distribution of the number of republicans in the senate.

#### Problem 3(a)

First, we define a Bayesian model for each race. The prior for the difference $\theta$ between republicans and democtrats will be $N(\mu,\tau^2)$. Say before seeing poll data you have no idea who is going to win, what should $\mu$ be? How about $\tau$, should it be large or small? 

** Your answer here: **

We take the prior difference $\theta$ to be normally distributed $\theta \sim N(\mu, \tau^2)$. If we have no information about the race, $\mu$ would be the average republican-democrat point difference for all senate races in recent relevant history. $\tau$ would be the standard deviation of those differences. If we assume that republicans have won about just as often as they have lost then we can take $\mu\sim 0$. We also know that the point differences generally don't exceed 15 points so we could set the width to be at about $\tau \sim 15$. We will calculate these quantities exactly later. 

#### Problem 3(b)

What is the distribution of d conditioned on $\theta$. What is the posterior distribution of $\theta|d$?

**Hint**: Use normal approximation. 

**Your answer here:**

We take the observed difference $d$ conditioned on a state's intrinsic difference $\theta$ to be normally distributed: $d |\theta \sim N(\theta, \sigma^2)$. $\sigma$ is standard error of the difference among all polls for a state (this is to ensure that as the number of polls goes to infinite our distribution becomes tightly peaked at $\theta$.) The posterior distribution $\theta|d$ is also normally distributed. It has a mean and variance of the following forms.
$$ E(\theta|d) = \mu + \left(1-\frac{\sigma^2}{\sigma^2+\tau^2}\right)(d-\mu) $$
$$ Var(\theta|d) = \frac{1}{1/\sigma^2+1/\tau^2}$$

The posterior mean difference for each state is equivalent to the epected (as in, election day) R-D point difference for each state. To calculate the expected R-D point difference $\theta_k$ for a state $k$ we apply the formula
$$ E(\theta_k|d_k) = \mu + \left(1-\frac{\sigma_k^2}{\sigma_k^2+\tau^2}\right)(d_k-\mu) $$
where $d_k$ is the difference computed from the poll average, $\mu$ is the average difference among all states, $\sigma_k$ is the standard error of the state's polls, and $\tau$ is the standard deviation of all state differences. 

Thus we need to determine $d_k$ and $\sigma_k$ for each state. $\mu$ and $\tau$ for all states were determined above.

```python
## List of d_k and sigma_k for each state.

# We use the standard error because the standard error defines the width for the probability distribution of the intrinsic mean.

means = list()
st_errors = list()
for k in range(len(candidates)):
    Rep_can = candidates.ix[k]['R']
    Oth_can = candidates.ix[k]['D']
    Rep_percents = polls_date_cut[slug_list_cut[k]][Rep_can]
    Oth_percents = polls_date_cut[slug_list_cut[k]][Oth_can]
    diff = Rep_percents - Oth_percents
    
    mean_diff = float('%.2f' % np.mean(diff))
    ste_diff =  float('%.2f' % (np.std(diff)/np.sqrt(len(diff))) ) 

    st_errors.append(ste_diff)
    means.append(mean_diff)
## definition of expectation value and variance functions
```

```python
# expectation value function
def expec_value(dk, sigmak, m, t):
    E = m + (1- sigmak*sigmak/(sigmak*sigmak+t*t))*(dk-m)
    return float('%.2f' % E)

# variance function
def variance_value(sigmak, t):
    V =  (1/(sigmak*sigmak)+1/(t*t))**(-1)
    return float('%.2f' % V)
```

```python
# list of expected values conditioned on prior probability
# list of std values conditioned on prior probability

ex_means = list()
std_values = list()
for k in range(35):
    ex_means.append(expec_value(means[k],st_errors[k],mu, tau))
    std_values.append(np.sqrt(variance_value(st_errors[k], tau)))
```

```python
    
# plot of expected differences agains prior differences

plt.figure(figsize=(8,6))

plt.scatter(means, ex_means)

plt.title('Expected Differences vs. Poll Differences')

plt.xlabel("Poll Differences") # axis label
plt.ylabel("Expected Differences") # axis label

plt.show()
```

**Your answer here:**

There is hardly any difference (excuse the pun) between the expected differences and the poll differences. There is little difference because the standard deviation of R-D differences among all states is quite large this year $\sim 20$, while the variations between pollsters for each state is quite small $ ~3$. Since the variance of the measured distribution is much smaller than the variance of the prior distribution, the result we find is simply a consequence of the way Bayes' theorem tells us to ignore the data from the least precise distribution.
